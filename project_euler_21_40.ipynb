Let $d(n)$ be defined as the sum of proper divisors of $n$ (numbers less than $n$ which divide evenly into $n$). <br>
If $d(a) = b$ and $d(b) = a$ , where $a \ne b$, then $a$ and $b$ are an amicable pair and each of $a$ and $b$ are called amicable numbers.

For example, the proper divisors of $220$ are $1, 2, 4, 5, 10, 11, 20, 22, 44, 55$ and $110$; therefore $d(220) = 284$ . The proper divisors of $284$ are $1, 2, 4, 71$ and $142$; so $d(284) = 220$ .

Evaluate the sum of all the amicable numbers under $10000$.
</span>

```python
# let's import a library that returns the divisors of a number

from sympy import divisors
```

```python
def is_amicable(a):
    b = sum(divisors(a)[:-1])
    if a != b and sum(divisors(b)[:-1]) == a:
        return a
    else:
        return False
```

```python
def amicables(n):
    amicables = []
    i = 0
    while (i <= n) and (i not in amicables):
        if (a := is_amicable(i)):
            amicables.append(a)
        i += 1
    return amicables
```

```python
sum(amicables(10000))
```

<span style="font-size:20px">
Using names.txt (right click and 'Save Link/Target As...'), a 46K text file containing over five-thousand first names, begin by sorting it into alphabetical order. Then working out the alphabetical value for each name, multiply this value by its alphabetical position in the list to obtain a name score.<br>

For example, when the list is sorted into alphabetical order, COLIN, which is worth  $3 + 15 + 12 + 9 + 14 = 53$, is the $938$ th name in the list. So, COLIN would obtain a score of $938 \times 53 = 49714$ .

What is the total of all the name scores in the file?
</span>

```python
import ast #Abstract Syntax Tree - a built-in Python module for safely evaluating Python expressions

with open('/Users/tanya/Desktop/0022_names.txt', 'r') as file:
    content = file.read()
    names_list = ast.literal_eval(f"[{content}]")
    names_list.sort()
    print(names_list[:10])
```

```python
abc = {
    'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6, 'g': 7, 'h': 8, 'i': 9, 'j': 10,
    'k': 11, 'l': 12, 'm': 13, 'n': 14, 'o': 15, 'p': 16, 'q': 17, 'r': 18, 's': 19,
    't': 20, 'u': 21, 'v': 22, 'w': 23, 'x': 24, 'y': 25, 'z': 26
}
```

```python
def letter_score(name, index):
    alph_value = 0
    letters = list(name.lower())
    for i in letters:
        alph_value += abc[i]
    score = alph_value * index
    return score
```

```python
total_score = 0
for idx, name in enumerate(names_list):
    score = letter_score(name, idx+1)
    total_score += score

print(total_score)
```
