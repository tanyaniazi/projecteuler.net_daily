# Problem 61

Triangle, square, pentagonal, hexagonal, heptagonal, and octagonal numbers are all figurate (polygonal) numbers and are generated by the following formulae:

| Type | Formula | First terms |
| --- | --- | --- |
| Triangle | $P_{3,n}=n(n+1)/2$ | $1, 3, 6, 10, 15, \dots$ |
| Square | $P_{4,n}=n^2$ | $1, 4, 9, 16, 25, \dots$ |
| Pentagonal | $P_{5,n}=n(3n-1)/2$ | $1, 5, 12, 22, 35, \dots$ |
| Hexagonal | $P_{6,n}=n(2n-1)$ | $1, 6, 15, 28, 45, \dots$ |
| Heptagonal | $P_{7,n}=n(5n-3)/2$ | $1, 7, 18, 34, 55, \dots$ |
| Octagonal | $P_{8,n}=n(3n-2)$ | $1, 8, 21, 40, 65, \dots$ |

The ordered set of three 4-digit numbers, $8128$, $2882$, and $8281$, has three interesting properties:

1. The set is cyclic, in that the last two digits of each number are the first two digits of the next number, including the last number with the first.
2. Each polygonal type is represented by a different number in the set: triangle ($P_{3,127}=8128$), square ($P_{4,91}=8281$), and pentagonal ($P_{5,44}=2882$).
3. This is the only set of 4-digit numbers with this property.

Find the sum of the only ordered set of six cyclic 4-digit numbers for which each polygonal type, triangle, square, pentagonal, hexagonal, heptagonal, and octagonal, is represented by a different number in the set.


```python
import math
import sympy
```

```python
def is_triangle(i):
    if i < 0: return False
    val = 1 + 8 * i
    root = math.isqrt(val)
    if root * root == val:
        return (-1 + root) % 2 == 0
    return False

def is_square(i):
    if i < 0: return False
    root = math.isqrt(i)
    return root * root == i

def is_pentagonal(i):
    if i <= 0:
        return False
    val = 1 + 24 * i
    root = math.isqrt(val)
    if root * root == val:
        return (1 + root) % 6 == 0
    return False
```

```python
def is_hexagonal(i):
    val = 1 + 8 * i
    root = math.isqrt(val)
    if root * root == val:
        return (1 + root) % 4 == 0
    return False

def is_heptagonal(i):
    val = 9 + 40 * i
    root = math.isqrt(val)
    if root * root == val:
        return (3 + root) % 10 == 0
    return False

def is_octagonal(i):
    val = 4 + 12 * i
    root = math.isqrt(val)
    if root * root == val:
        return (2 + root) % 6 == 0
    return False
```

```python
triangles = []
squares = []
pentagonals = []
hexagonals = []
heptagonals = []
octagonals = []


for i in range(1000, 10000):
    if is_triangle(i):
        triangles.append(i)
    if is_square(i):
        squares.append(i)
    if is_pentagonal(i):
        pentagonals.append(i)
    if is_hexagonal(i):
        hexagonals.append(i)
    if is_heptagonal(i):
        heptagonals.append(i)
    if is_octagonal(i):
        octagonals.append(i)
```

```python
sets = [squares, pentagonals, hexagonals, heptagonals, octagonals]
```

```python
def find_chain(last_num, remaining_sets, chain):
    if not remaining_sets:
        if str(chain[-1])[2:] == str(chain[0])[:2]:
            print("Found it!", chain, "Sum:", sum(chain))
            return True
        return False

    suffix = str(last_num)[2:]
    
    for i, current_set in enumerate(remaining_sets):
        for num in current_set:
            if str(num).startswith(suffix):
                new_sets = remaining_sets[:i] + remaining_sets[i+1:]
                if find_chain(num, new_sets, chain + [num]):
                    return True
    return False

for tri in triangles:
    if find_chain(tri, sets, [tri]):
        break
```
